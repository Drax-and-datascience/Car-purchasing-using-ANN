Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

Loading the dataset, removing irrelevant columns, setting features and targets

```python
dataset = pd.read_csv('car_purchasing.csv', encoding='ISO-8859-1')
dataset= dataset.drop(columns =['customer name', 'customer e-mail', 'country'])
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
dataset.describe(include = 'all')
```

Train test and split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Feature scaling ( Do not fporget to scale dependent variable as well)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
sc1 = StandardScaler()
y_train = sc1.fit_transform(y_train)
```

Loading and training the model (try experimenting with the no. of units in hidden layer, no. of epochs, batch size)

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=50, activation='relu', input_dim = 5))
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
ann.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = 'mean_absolute_error')
ann.fit(X_train, y_train,  epochs = 100, validation_split=0.4)
```

```python
y_pred = ann.predict(X_test)
y_pred = sc1.inverse_transform(y_pred)
y_pred
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_test, y_pred)
print("R2 Score=",R2 )
```
